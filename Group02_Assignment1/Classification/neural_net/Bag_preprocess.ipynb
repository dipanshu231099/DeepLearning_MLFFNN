```python
f = open("image_test.txt")

X = []
Y = []
for line in f:
    temp  = line.split()
    temp = [float(i) for i in temp]
    temp = np.array(temp)
    X.append(temp)
f = open("image_test_label.txt")    
for line in f:
    temp  = line.split()
    temp = [float(i) for i in temp]
    temp = np.array(temp)
    Y.append(temp[0])
Y = np.array(Y)
X = np.array(X)

class1 = []
class2 = []
class3 = []
for i in range(len(X)):
    if(Y[i]==0):
        class1.append(X[i])
    elif(Y[i]==1):
        class2.append(X[i])
    else:
        class3.append(X[i])
class1 = np.array(class1)
class2 = np.array(class2)
class3 = np.array(class3)
np.savetxt('c1_test.txt', np.array(class1))
np.savetxt('c2_test.txt', np.array(class2))
np.savetxt('c3_.test.txt', np.array(class3))
```
